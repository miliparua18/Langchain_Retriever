```python
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document
```

```python
#Source documents
documents=[
    Document(page_content="Lnagchain helps to build a llm based applications"),
    Document(page_content="Chroma is vector database  optimized for llm based search"),
    Document(page_content="Embedding helps to convers the texts in vectors"),
    Document(page_content="Huggingface provide powerful embedding models")
]
```

```python
embedding_model = HuggingFaceEmbeddings()

#create vector_store uding chroma model
Vector_store = Chroma.from_documents(
    documents = documents,
    embedding=embedding_model,
)
```

```python
retriever = Vector_store.as_retriever(
    search_type='mmr',
    search_kwargs= {"k":1,"lambda_mult":1}
)
```

```python
query="What is langchain?"
result = retriever.invoke(query)
```

```python
#print retrivers content
for i,result in enumerate(result):
  print(f"\n---Result {i+1} ---")
  print(f"content: \n{result.page_content}--- ")
```
