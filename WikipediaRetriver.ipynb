```python
from langchain_community.retrievers import WikipediaRetriever
```

```python
#initialize the retrivers(optional: top_k_results, lang)

retrievers = WikipediaRetriever(top_k_results=2, lang='en')
```

```python
#define query
query = "The geopolitical history of india and pakistan from the perspective of a chinese"
docs = retrievers.invoke(query)
print(docs)
```

```python
#print retrivers content
for i,docs in enumerate(docs):
  print(f"\n---Result {i+1} ---")
  print(f"content: \n{docs.page_content}--- ")  #truncate for display
```
