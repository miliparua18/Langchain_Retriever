```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEmbeddings
from langchain_core.documents import Document
from langchain_community.vectorstores import Chroma
from langchain_community.retrievers import ContextualCompressionRetriever
from langchain_community.retrievers import LLMChainExtractor
```

```python
contextual_documents = [
    Document(
        page_content="Ravi recently started learning Python for data science. He found it easy to understand because of its simple syntax. Now, he is working on a machine learning project using libraries like pandas and scikit-learn.",
        metadata={"topic": "Programming Story", "level": "basic"}
    ),

    Document(
        page_content="A small village in India has started using solar panels to generate electricity. This initiative has reduced power cuts and improved the quality of life for residents.",
        metadata={"topic": "Environment", "level": "intermediate"}
    ),

    Document(
        page_content="During the COVID-19 pandemic, many people shifted to online learning platforms. This change helped students continue their education but also highlighted the digital divide.",
        metadata={"topic": "Education", "level": "intermediate"}
    ),

    Document(
        page_content="Neha wanted to improve her fitness, so she began jogging every morning and eating a balanced diet. After a few months, she noticed significant improvements in her health and energy levels.",
        metadata={"topic": "Health Story", "level": "basic"}
    ),

    Document(
        page_content="Space agencies like NASA and ISRO are planning future missions to Mars. These missions aim to study the planet's environment and search for signs of past life.",
        metadata={"topic": "Space", "level": "intermediate"}
    )
]
```

```python
embedding_model = ChatHuggingFace()
vector_store = Chroma.from_documents(contextual_documents,embedding_model)
```

```python
base_retriever = vector_store.as_retriever(search_kwargs={"k":5})
```

```python
#set up the compressor using llm

llm = ChatHuggingFace(model='openai/gpt-oss-20b')
compressor = LLMChainExtractor.from_llm(llm)
```

```python
compressor_retriever = ContextualCompressionRetriever(
    base_compressor = compressor,
    base_retriever = base_retriever
)
```

```python
query = 'What happend during COVID-19 pandamic?'
result = compressor_retriever.invoke(query)
```
