```python
from langchain_huggingface import HuggingFaceEmbeddings, ChatHuggingFace
```

```python
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
documents = [
     Document(
        page_content="Diabetes is a chronic disease where the body cannot properly regulate blood sugar levels. It can be managed with diet, exercise, and medication.",
        metadata={"topic": "Diabetes", "level": "basic"}
    ),

    Document(
        page_content="Hypertension, or high blood pressure, increases the risk of heart disease and stroke. Lifestyle changes and medication can help control it.",
        metadata={"topic": "Hypertension", "level": "intermediate"}
    ),

    Document(
        page_content="A balanced diet includes fruits, vegetables, proteins, and whole grains. It provides essential nutrients needed for maintaining good health.",
        metadata={"topic": "Nutrition", "level": "basic"}
    ),

    Document(
        page_content="Regular exercise improves heart health, builds strength, and reduces stress. Walking, jogging, and yoga are simple ways to stay active.",
        metadata={"topic": "Fitness", "level": "basic"}
    ),

    Document(
        page_content="Mental health includes emotional, psychological, and social well-being. Managing stress and seeking support are important for mental wellness.",
        metadata={"topic": "Mental Health", "level": "basic"}
    ),

      Document(
        page_content="Python is a high-level programming language used in web development, data science, artificial intelligence, and automation.",
        metadata={"topic": "Programming", "level": "basic"}
    ),

    Document(
        page_content="Space exploration helps scientists understand planets, stars, and galaxies. Organizations like NASA continue to explore space.",
        metadata={"topic": "Space", "level": "intermediate"}
    ),

    Document(
        page_content="Online learning platforms allow students to study from anywhere. They offer flexibility and access to a wide range of courses.",
        metadata={"topic": "Education", "level": "basic"}
    ),


]
```

```python
embedding_model = HuggingFaceEmbeddings()

#create vector_store uding chroma model
Vector_store = Chroma.from_documents(
    documents = documents,
    embedding=embedding_model,
)
```

```python
from re import search
similarity_retriever = Vector_store.as_retriever(search_type="similarity",search_kwargs={'k':3})
```

```python
Multiquery_Retriever = MultiQueryRetriever.from_llm(
   retriever = Vector_store.as_retriever(search_kwargs={'k':1}),
   llm=ChatHuggingFace(model="openai/gpt-oss-20b")
)
```

```python
query="How can i inprove my energy level"
```

```python
result = Multiquery_Retriever.invoke(query)
```

```python
#print retrivers content
for i,result in enumerate(result):
  print(f"\n---Result {i+1} ---")
  print(f"content: \n{result.page_content}--- ")
```
